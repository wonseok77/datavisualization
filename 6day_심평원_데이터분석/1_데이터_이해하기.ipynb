### 메타정의서 eng & kor 칼럼명 가지고 오기

```python
file_path = './files_org/표본자료 변수 테이블(2012-2013).xlsx'

## excel 파일 불러올때 sheet는 default로 0번재 sheet를 불러온다
## sheet_name 매개변수로 번호를 입력하여 사용할수있고 ''안에 시트이름을 입력하여 사용할수도 있다
## 번호를 사용하는 방법은 나중에 헷갈릴 가능성이 많다

df_20TABLE_col = pd.read_excel(file_path,
                               sheet_name = '20table_명세서일반내역',
                               usecols='A,D')
df_20TABLE_col.head(1)
```

```python
df_30TABLE_col = pd.read_excel(file_path,
                               sheet_name = '30table_진료내역',
                               usecols='A,D')
df_30TABLE_col.head(1)
```

```python
df_40TABLE_col = pd.read_excel(file_path,
                               sheet_name = '40table_상병',
                               usecols='A,D')
df_40TABLE_col.head(1)
```

```python
df_53TABLE_col = pd.read_excel(file_path,
                               sheet_name = '53table_처방전상세내역',
                               usecols='A,D')
df_53TABLE_col.head(1)
```

```python
'''
데이터프레임의 컬럼명 == 딕셔너리의 key
    딕셔너리의 value값으로 대체
'''
# temp_dict = {}
# temp_dict["key1"] = 'value1'
# temp_dict["key2"] = 'value2'
# temp_dict["key3"] = 'value3'
# temp_dict
```

```python
df_20TABLE_col_list = df_20TABLE_col.to_dict("records")
new_dict_20 = {}

for dict_kv in df_20TABLE_col_list:
    # print(dict_kv['변수명'],dict_kv['표준한글컬럼명'])
    new_dict_20[dict_kv['변수명']] = dict_kv['표준한글컬럼명']
df_20TABLE_kor = df_20TABLE.rename(columns = new_dict_20)
df_30TABLE_kor = df_30TABLE.rename(columns = new_dict_20)
df_40TABLE_kor = df_40TABLE.rename(columns = new_dict_20)
df_53TABLE_kor = df_53TABLE.rename(columns = new_dict_20)
```
